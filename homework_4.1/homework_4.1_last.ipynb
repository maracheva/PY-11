```python
import pandas as pd
import os
```

```python
original_dir = 'names'  # папка с исходными файлами
DATA_PATH = original_dir # 'C:\\names' 
```

```python
# Функция возвращает набор данных [имя, пол, кол.] по годам.
def all_years(year):
    all_years = {}
    for y in year:
        all_years[y] = pd.read_csv(f'{DATA_PATH}/yob{y}.txt', names=['Name','Pol','Count']) 
        all_names = pd.concat(all_years, names=['Year', 'Pos'])
    return all_names
```

```python
all_years([1980])
```

```python
# Функция загружает указанные года и выводит ТОП-3 популярных имен. 
def get_top3(years):
    all_names = all_years(years)
    for year in years:
# отсортируем по кол-ву и выведем список ТОП-3 самых популярных имен
        group_names = all_names.groupby('Name').Count.sum().to_frame()
        list_top_names = group_names.sort_values('Count', ascending=False).head(3).index.tolist()
    return list_top_names
```

```python
get_top3([1880])
```

```python
get_top3([1998, 1974])
```

```python
# Функция возвращает динамику изменения количества имен за указанные года в разрезе полов.
def dynamic_names(years):
    quantity_names = {}
    group_years = all_years(years)
    # сгруппируем данные по Year, суммируем кол-во данного Year
    dynamic_names_female = group_years[group_years.Pol=='F'].groupby('Year').Count.sum().to_frame()['Count']
    dynamic_names_male = group_years[group_years.Pol=='M'].groupby('Year').Count.sum().to_frame()['Count']
  
    # выведем в список
    quantity_names['F'] = dynamic_names_female.tolist()
    quantity_names['M'] = dynamic_names_male.tolist()
    return quantity_names
```

```python
dynamic_names([1900, 1950, 2000])
```
